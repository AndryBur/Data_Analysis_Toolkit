```python
#----------------------------------------- обработка текста --------------------------------------
# Эдгар По
text = '''
Ah, broken is the golden bowl!
The spirit flown forever!
Let the bell toll! - A saintly soul
Glides down the Stygian river!
And let the burial rite be read -
The funeral song be sung -
A dirge for the most lovely dead
That ever died so young!
And, Guy de Vere,
Hast thou no tear?
Weep now or nevermore!
See, on yon drear
And rigid bier,
Low lies thy love Lenore!

"Yon heir, whose cheeks of pallid hue
With tears are streaming wet,
Sees only, through
Their crocodile dew,
A vacant coronet -
False friends! ye loved her for her wealth
And hated her for her pride,
And, when she fell in feeble health,
Ye blessed her - that she died.
How shall the ritual, then, be read?
The requiem how be sung
For her most wrong'd of all the dead
That ever died so young?"

Peccavimus!
But rave not thus!
And let the solemn song
Go up to God so mournfully that she may feel no wrong!
The sweet Lenore
Hath "gone before"
With young hope at her side,
And thou art wild
Fot the dear child
That should have been thy bride -
For her, the fair
And debonair,
That now so lowly lies -
The life still there
Upon her hair,
The death upon her eyes.

"Avaunt! - to-night
My heart is light -
No dirge will I upraise,
But waft the angel on her flight
With a Pæan of old days!
Let no bell toll!
Lest her sweet soul,
Amid its hallow'd mirth,
Should catch the note
As it doth float
Up from the damned earth -
To friends above, from fiends below,
th' indignant ghost is riven -
From grief and moan
To a gold throne
Beside the King of Heaven!"
'''
```

```python
text = text.replace('\n', ' ').replace('"','').replace('.',' ').replace('-',' ').replace(',',' ').replace('!',' ').replace('?',' ').replace('  ', ' ').lower()
text_splitted = text.split(' ')
while '' in text_splitted:
    text_splitted.remove('')
len(text_splitted)
```

```python
vocabulary = set(text_splitted)
len(vocabulary)
```

```python
frequency_dictionary = {word:text_splitted.count(word) for word in vocabulary}
frequency_dictionary
```

```python
temp_dict = frequency_dictionary.copy()

ordered_list_of_words = []

while temp_dict:
    max_num = max(temp_dict.values())
    keys = temp_dict.keys()
    for word in keys:
        if temp_dict[word] == max_num:
            current_word = word
            break
    temp_dict.pop(current_word)
    ordered_list_of_words.append(current_word)

ordered_list_of_words
```

```python
nltk.download('wordnet')
Lemmatizer = WordNetLemmatizer()

text_splitted_lemmatize = [Lemmatizer.lemmatize(word) for word in text_splitted]
vocabulary_lemmatize = set(text_splitted_lemmatize)
len(vocabulary_lemmatize)
```

```python
x = ordered_list_of_words.copy()
y = [frequency_dictionary[word] for word in x]

x1 = [i for i in range(len(y))]

plt.figure(figsize=(15, 5))
plt.grid()
plt.bar(x1,y)
```

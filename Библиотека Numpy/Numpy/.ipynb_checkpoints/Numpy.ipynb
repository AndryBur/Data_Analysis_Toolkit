```python
import matplotlib.pyplot as plt
image = plt.imread("Балкон.jpg")
plt.figure(figsize=(15, 13))
plt.imshow(image)
```

```python
image.shape
```

```python
image_transposed = np.transpose(image, (1, 0, 2))
```

```python
# транспонирование (поворот) изображениея
plt.figure(figsize=(12, 10))
plt.imshow(image_transposed)
```

```python
# инверсия цветов
plt.figure(figsize=(15, 13))
plt.imshow(image[:, :, ::-1])
```

```python
# разрежение изображения
plt.figure(figsize=(15, 13))
plt.imshow(image[::4, ::4])
```

```python
part_of_image = image[:190, 50:290, :]
plt.figure(figsize=(8, 6))
plt.imshow(part_of_image)
```

```python
lustra = np.array(image)
lustra[:190, 45:50, 0] = 0
lustra[:190, 295:300, 0] = 0
lustra[:190, 45:50, 1] = 255
lustra[:190, 295:300, 1] = 255
lustra[:190, 45:50, 2] = 0
lustra[:190, 295:300, 2] = 0

lustra[190:195, 45:300, 0] = 0
lustra[190:195, 45:300, 1] = 255
lustra[190:195, 45:300, 2] = 0


plt.figure(figsize=(15, 13))
plt.imshow(lustra)
```

```python
#---------------------------------------- генарация случайного распределения --------------------------------------------------

#---------------------------------------- нормальное распределения с центром в 0 и единичной дисперсией --------------------------------------------------

normal = random.normal(size=1000000)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(5, 5))
plt.grid()
hist = plt.hist(normal, bins=2000)
```

```python
mu = 10
normal = random.normal(loc=mu, size=1000000)
plt.figure(figsize=(5, 5))
plt.grid()
hist = plt.hist(normal, bins=200)
```

```python
sigma = 10
normal = random.normal(scale=sigma, size=1000000)
plt.figure(figsize=(5, 5))
plt.grid()
hist = plt.hist(normal, bins=200)
```

```python
mu = 15_000
sigma = 10
normal = random.normal(loc=mu, scale=sigma, size=1000000)
plt.figure(figsize=(5, 5))
plt.grid()
hist = plt.hist(normal, bins=200)
```

```python
mu = np.array([10, 100, 1000]).reshape(-1, 1)
sigma = np.array([3, 30, 300]).reshape(-1, 1)
normal = random.normal(loc=mu, scale=sigma, size=(3, 100000))
plt.figure(figsize=(5, 5))
plt.grid()
hist = plt.hist(normal[0,:], bins=200)
```

```python
#---------------------------------------- юниформ-распределения --------------------------------------------------
uniform = random.uniform(size=10000)
plt.figure(figsize=(5, 5))
plt.grid()
hist = plt.hist(uniform, bins=200)
```
